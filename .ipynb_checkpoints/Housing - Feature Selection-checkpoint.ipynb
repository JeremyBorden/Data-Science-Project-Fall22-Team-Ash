### Now trying some lasso regression on these features
And their polynomial transformations - 'YrAfterMinBuilt' really seems the only one to me amenable to maybe a parabola, but not a huge computational burden to also check the squares of the other terms. Plus polynomial features will let us check interactions too.

```python
# Some code taken from Matt's lectures
alphas = [0.000001,.00001,.0001,.001,.01,.1,.25,.5,.75,1]

# Get polynomial features up to degree 2
poly = PolynomialFeatures(2)
houses_poly = poly.fit_transform(houses_temp[new_num])

# Scale all the features
scaler = StandardScaler()
houses_poly_scaled = scaler.fit_transform(houses_poly)

# Set up an array to hold the lasso coefficients
coefs = np.zeros((len(alphas), len(houses_poly_scaled[1,:])) )

## Loop through alphas
for i in range(len(alphas)):
    ## make lasso model
    lasso = Lasso(alpha=alphas[i], max_iter=10000000)
    
    
    ## fit model
    lasso.fit(houses_poly_scaled, houses_temp['LogSalePrice'])
    
    ## record coefs
    coefs[i,:] = lasso.coef_
# get_feature_names_out will give us nice names for the polynomial features
cols = poly.get_feature_names_out(houses_temp[new_num].columns)
lasso_coefs = pd.DataFrame(coefs, index=alphas, columns = cols)
lasso_coefs
```

```python
# Take the columns with nonzero coefficients in row 7 (alpha = 0.1) as barebones model (this is the last row
# with any nonzero coefficients)
cut = 5
print(lasso_coefs.index[cut])
for col in lasso_coefs.columns:
    if lasso_coefs[col].iloc[cut] != 0:
        print(col)
```

## Now try to narrow down to one of these subsets with regression models and Kfold cross validation

```python
# Make all the additional columns we need

houses_temp['YrAfterMinBuilt_GrLivArea'] = houses_temp['YrAfterMinBuilt']*houses_temp['GrLivArea']
houses_temp['YrAfterMinRemod_GrLivArea'] = houses_temp['YrAfterMinRemod']*houses_temp['GrLivArea']
houses_temp['YrAfterMinRemod_GarageArea'] = houses_temp['YrAfterMinRemod']*houses_temp['GarageArea']
houses_temp['YrAfterMinBuilt_1stFlrSF'] = houses_temp['YrAfterMinBuilt']*houses_temp['1stFlrSF']
houses_temp['YrAfterMinBuilt_YrAfterMinRemod'] = houses_temp['YrAfterMinBuilt']*houses_temp['YrAfterMinRemod']
houses_temp['YrAfterMinBuilt_GarageArea'] = houses_temp['YrAfterMinBuilt']*houses_temp['GarageArea']
houses_temp['TotalBsmtSFSquared'] = houses_temp['TotalBsmtSF']**2

houses_temp['LogYrAfterMinBuilt_LogGrLivArea'] = houses_temp['LogYrAfterMinBuilt']*houses_temp['LogGrLivArea']
houses_temp['Log1stFlrSF_LogGrLivArea'] = houses_temp['Log1stFlrSF']*houses_temp['LogGrLivArea']
houses_temp['LogLotArea_LogGrLivArea'] = houses_temp['LogLotArea']*houses_temp['LogGrLivArea']
houses_temp['LogYrAfterMinRemod_LogTotalBsmtSF'] = houses_temp['LogYrAfterMinRemod']*houses_temp['LogTotalBsmtSF']
houses_temp['LogYrAfterMinRemod_LogGarageArea'] = houses_temp['LogYrAfterMinRemod']*houses_temp['LogGarageArea']
houses_temp['LogGrLivAreaSquared'] = houses_temp['LogGrLivArea']**2
houses_temp['LogYrAfterMinBuiltSquared'] = houses_temp['LogYrAfterMinBuilt']**2
houses_temp['LogYrAfterMinRemodSquared'] = houses_temp['LogYrAfterMinRemod']**2
houses_temp['LogTotalBsmtSFSquared'] = houses_temp['LogTotalBsmtSF']**2
houses_temp['LogTotalBsmtSF_LogGarageArea'] = houses_temp['LogTotalBsmtSF']*houses_temp['LogGarageArea']
houses_temp['LogGarageAreaSquared'] = houses_temp['LogGarageArea']**2
```

```python
kfold = KFold(n_splits = 5,
                 shuffle = True,
                 random_state = 214)
```

```python
# List with all 6 different numerical feature subsets from the lasso regression
models = [lasso_bare, lasso_cons, lasso_gen, lasso_log_bare, lasso_log_cons, lasso_log_gen]

# Arrays to hold rmses and r2_scores (rows are models, columns are each cross validation of the kfold)
rmses = np.zeros((7, 5))
r2_scores = np.zeros((7, 5))
adj_r2_scores = np.zeros((7, 5))

i = 0
for train_index, test_index in kfold.split(houses_temp):
    
    # Get the train-train and holdout sets
    houses_tt = houses_temp.iloc[train_index]
    houses_ho = houses_temp.iloc[test_index]
    
    # Baseline model (just the mean of 'SalePrice')
    avg = houses_tt['SalePrice'].mean()*np.ones(len(houses_ho))
    rmses[0,i] = np.sqrt(mse(houses_ho['SalePrice'], avg ))
    r2_scores[0,i] = r2_score(houses_ho['SalePrice'], avg )
    adj_r2_scores[0,i] = 1 - ( (1-r2_score(houses_ho['SalePrice'], avg ))*((len(houses_tt)-1)/(len(houses_tt)-0) ) )
    
    # Loop through the 6 different feature subsets
    j = 1
    for feat_list in models:
        # Fit the scaler to the training set and scale the relevant columns of the training and holdout sets
        scaler = StandardScaler()
        scaler.fit(houses_tt[feat_list])
        houses_tt_scaled = scaler.transform(houses_tt[feat_list])
        houses_ho_scaled = scaler.transform(houses_ho[feat_list])
        
        # Fit the regression in all the (scaled) features of the model
        reg = LinearRegression()
        reg.fit(houses_tt_scaled, houses_tt['LogSalePrice'])
        pred = reg.predict(houses_ho_scaled)
        pred = np.power(10, pred)
        rmses[j,i] = np.sqrt(mse(houses_ho['SalePrice'], pred))
        r2_scores[j,i] = r2_score(houses_ho['SalePrice'], pred)
        adj_r2_scores[j,i] = 1 - ( (1-r2_score(houses_ho['SalePrice'], pred))*((len(houses_tt)-1)/(len(houses_tt)-len(feat_list)) ) )
        j = j + 1
    i = i + 1
avg_kfold_rmses = rmses.mean(axis=1)
avg_kfold_r2_scores = r2_scores.mean(axis=1)
avg_kfold_adj_r2_scores = adj_r2_scores.mean(axis=1)
```

```python
avg_kfold_rmses.argmin()
```

```python
avg_kfold_r2_scores.argmax()
```

```python
avg_kfold_adj_r2_scores.argmax()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
fig.set_figheight(6)
fig.set_figwidth(16)
for i in range(7):
    ax1.scatter(avg_kfold_rmses[i], avg_kfold_r2_scores[i], s=60, label = 'Model '+str(i))
ax1.set_xlabel('Average kfold rmse', fontsize = 16)
ax1.set_ylabel('Average kfold R^2 score', fontsize = 16)
ax1.legend(fontsize = 16)

for i in range(7):
    ax2.scatter(avg_kfold_rmses[i], avg_kfold_adj_r2_scores[i], s=60, label = 'Model '+str(i))
ax2.set_xlabel('Average kfold rmse', fontsize = 16)
ax2.set_ylabel('Average kfold adjusted R^2 score', fontsize = 16)
ax2.legend(fontsize = 16)




plt.show()
```

## Results

Model 0 is the baseline model (simply taking the mean of 'SalePrice' in the training set). At least a moderately good sign is that every model outperformed the baseline in terms of the RMSE. 

All the average adjusted R^2 scores were very close to the regular R^2 scores (there weren't all that many features - the largest set was 16 features - and in the houses_tt set there were ~960 observations.

That being said, the winner here, both in terms of highest R^2 (and adjusted R^2) score and lowest RMSE is model 3 ('lasso_gen', the generous lasso model). 

```python
print('lasso_gen avg kfold RMSE:', avg_kfold_rmses[3])
print('lasso_gen avg kfold R^2 score:', avg_kfold_r2_scores[3])
print('lasso_gen avg kfold adjusted R^2 score:', avg_kfold_adj_r2_scores[3])
```

```python
# Function to get dummy variables for a list of features in a dataframe
def dummies_subset(feat_list, df):
    # feat_list is the list of features you want dummy variables for
    # df is the data frame
    dummies = pd.DataFrame()
    for feature in feat_list:
        temp = pd.DataFrame()
        for col in pd.get_dummies(df[feature]).columns[:]:
            # omit last class in each category (whichever pandas puts last, that is)
            new_label = feature + '_' + str(col)
            temp[new_label] = pd.get_dummies(df[feature])[col]
        dummies = pd.concat([dummies, temp], axis = 1 )
    return dummies
```

```python
alphas = [0.000001,.00001,.0001,.001,.01,.025,.05,.1,.25,.5,.75,1]

## make coefficient holder
coefs = np.zeros((len(alphas), len(dummies.columns)) )

## Loop through alphas
for i in range(len(alphas)):
    ## make lasso model
    lasso = Lasso(alpha=alphas[i], max_iter=10000000)
    
    
    ## fit model
    lasso.fit(dummies, houses_temp_cat['LogSalePrice'])
    
    ## record coefs
    coefs[i,:] = lasso.coef_
```

## Explanation of the problem I was facing:
I was doing my one-hot encoding using the Pandas function get_dummies. However, several categorical variables have some categories with very few observations. Thus when I generated a training and test set, I sometimes didn't have all the categories for each variable in both sets. Then using the function get_dummies, I have arrays of features in my train/test set which are different shapes (and so incompatible, leaving me unable to generate any predictions).

### tldr: I've learned I should use sklearn's 'OneHotEncoder' object rather than the pandas function get_dummies, as 'OneHotEncoder' is compatible with some categories not being in the train or test set (it just ignores any such categories with the parameter handle_unknown = 'ignore')

```python
# Conservative and general lists of categorical features
print(lasso_cat_cons)
print(lasso_cat_gen)
```

```python
# Demonstration for my own sake of what ColumnTransformer and OneHotEncoder are doing
ct = ColumnTransformer([('one-hot-encoder', OneHotEncoder(drop='first'), lasso_cat_cons )], 
                           remainder = 'passthrough')

ohe_arr = ct.fit_transform(houses_temp_cat_drops)
ohe_arr
```

```python
# Now I'll try a simple linear regression model for both the conservative and generous feature lists to see if either 
# performs poorly enough to rule out for model building
```

```python
# Make a kfold
kfold = KFold(n_splits = 5,
                 shuffle = True,
                 random_state = 8989)
```

```python
# Arrays to hold rmses and r2_scores (rows are models, columns are each cross validation of the kfold)
rmses = np.zeros((3, 5))
r2_scores = np.zeros((3, 5))
adj_r2_scores = np.zeros((3, 5))

i = 0
for train_index, test_index in kfold.split(houses_temp_cat):
    
    # Get the train-train and holdout sets
    houses_tt = houses_temp_cat.iloc[train_index]
    houses_ho = houses_temp_cat.iloc[test_index]
    
    # Baseline model (just the mean of 'SalePrice'): Model 0
    avg = houses_tt['SalePrice'].mean()*np.ones(len(houses_ho))
    rmses[0,i] = np.sqrt(mse(houses_ho['SalePrice'], avg ))
    r2_scores[0,i] = r2_score(houses_ho['SalePrice'], avg )
    adj_r2_scores[0,i] = 1 - ( (1-r2_score(houses_ho['SalePrice'], avg ))*((len(houses_tt)-1)/(len(houses_tt)-0) ) )
    
    # Conservative subset of categorical features: Model 1
    
    # Make the ColumnTransformer with OneHotEncoder
    # Remainder set to 'drop' because I don't care about any features other than those I'm encoding (right now)
    ct_cons = ColumnTransformer([('one-hot-encoder', OneHotEncoder(drop='first',handle_unknown='ignore'), 
                                  lasso_cat_cons )], remainder = 'drop')
    ohe_arr_tt_cons = ct_cons.fit_transform(houses_tt)
    ohe_arr_ho_cons = ct_cons.transform(houses_ho)
    
    # Linear Regression
    reg_cons = LinearRegression()
    reg_cons.fit(ohe_arr_tt_cons, houses_tt['LogSalePrice'])
    pred_cons = reg_cons.predict(ohe_arr_ho_cons)
    pred_cons = np.power(10, pred_cons)
    
    # Get the performance metrics
    rmses[1,i] = np.sqrt(mse(houses_ho['SalePrice'], pred_cons ))
    r2_scores[1,i] = r2_score(houses_ho['SalePrice'], pred_cons )
    adj_r2_scores[1,i] = 1 - ( (1-r2_score(houses_ho['SalePrice'],pred_cons))*((len(houses_tt)-1)/(len(houses_tt)-ohe_arr_tt_cons.shape[1]) ) )
    

    # Same story with generous subset of categorical features: Model 2
    
    # Make the ColumnTransformer with OneHotEncoder
    # Remainder set to 'drop' because I don't care about any features other than those I'm encoding (right now)
    ct_gen = ColumnTransformer([('one-hot-encoder', OneHotEncoder(drop='first',handle_unknown='ignore'), 
                                 lasso_cat_gen )], remainder = 'drop')
    ohe_arr_tt_gen = ct_gen.fit_transform(houses_tt)
    ohe_arr_ho_gen = ct_gen.transform(houses_ho)
    
    # Linear Regression
    reg_gen = LinearRegression()
    reg_gen.fit(ohe_arr_tt_gen, houses_tt['LogSalePrice'])
    pred_gen = reg_gen.predict(ohe_arr_ho_gen)
    pred_gen = np.power(10, pred_gen)
    
    # Get the performance metrics
    rmses[2,i] = np.sqrt(mse(houses_ho['SalePrice'], pred_gen ))
    r2_scores[2,i] = r2_score(houses_ho['SalePrice'], pred_gen )
    adj_r2_scores[2,i] = 1 - ( (1-r2_score(houses_ho['SalePrice'],pred_gen))*((len(houses_tt)-1)/(len(houses_tt)-ohe_arr_tt_gen.shape[1]) ) )
    
    i = i + 1
avg_kfold_rmses = rmses.mean(axis=1)
avg_kfold_r2_scores = r2_scores.mean(axis=1)
avg_kfold_adj_r2_scores = adj_r2_scores.mean(axis=1)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2)
fig.set_figheight(6)
fig.set_figwidth(16)
for i in range(3):
    ax1.scatter(avg_kfold_rmses[i], avg_kfold_r2_scores[i], s=60, label = 'Model '+str(i))
ax1.set_xlabel('Average kfold rmse', fontsize = 16)
ax1.set_ylabel('Average kfold R^2 score', fontsize = 16)
ax1.legend(fontsize = 16)

for i in range(3):
    ax2.scatter(avg_kfold_rmses[i], avg_kfold_adj_r2_scores[i], s=60, label = 'Model '+str(i))
ax2.set_xlabel('Average kfold rmse', fontsize = 16)
ax2.set_ylabel('Average kfold adjusted R^2 score', fontsize = 16)
ax2.legend(fontsize = 16)




plt.show()
```

```python
# This is a function I wrote when I was still trying to work with get_dummies instead of OneHotEncoder
# Don't want to delete it just incase, so it's just in storage down here

# Function to get dummy variables for a list of features in a dataframe dropping whichever category has
# the fewest observations
def dummies_subset_drop_min(feat_list, df):
    # feat_list is the list of features you want dummy variables for
    # df is the data frame
    dummies = pd.DataFrame()
    for feature in feat_list:
        temp = pd.DataFrame()
        
        # first find category with the fewest counts
        min_index = df[feature].value_counts().argmin()
        min_cat = df[feature].value_counts().index[min_index]
        
        # Now loop through the dummy columns
        for col in pd.get_dummies(df[feature]).columns:
            # Make sure each column isn't the column with the fewest counts (min_cat)
            if col != min_cat:
                new_label = feature + '_' + str(col)
                temp[new_label] = pd.get_dummies(df[feature])[col]
        dummies = pd.concat([dummies, temp], axis = 1 )
    return dummies
```
